## Import data & descriptive statistical analysis

```python
data = pd.read_csv(r"/kaggle/input/air-passengers/AirPassengers.csv")
data['Month']  = pd.to_datetime(data['Month'])
data.set_index('Month',inplace=True)
data
```

## Data staratory checking, data diff and white noise checking

As can be seen from the data plot, the data is unstable as the years grow and has strong seasonality. We differentiate the data and plot the data

```python
import matplotlib.pyplot as plt

plt.plot(data)
```

diffenentive data and ADF checking

```python
data_diff1 = data.diff(1).dropna()
data_diff2 = data.diff(2).dropna()
data_diff3 = data.diff(3).dropna()
data_diff4 = data.diff(4).dropna()


plt.subplot(411)
plt.plot(data_diff1,"b")
plt.subplot(412)
plt.plot(data_diff2,"r")
plt.subplot(413)
plt.plot(data_diff3,"g")
plt.subplot(414)
plt.plot(data_diff4,"y")
plt.tight_layout()
```
